```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Line Graph

```python
gas = pd.read_csv('/content/gas_prices.csv')
gas
```

```python
plt.title('Gas price over time (in USD)')

plt.plot(gas.Year, gas.USA, label = 'USA')
plt.plot(gas.Year, gas.Canada, label = 'Canada')
plt.plot(gas.Year, gas['South Korea'], label = 'South Korea')

plt.xticks(gas.Year[:: 3])

plt.xlabel('Year')
plt. ylabel('Us Dollars')

plt.legend()
plt.show()
```

Graph all data

```python
for country in gas:
  if(country != 'Year'):
    plt.plot(gas.Year,gas[country],marker='.', label = country)

plt.legend(bbox_to_anchor=(1.1, 1.05)) # to side by side
```
